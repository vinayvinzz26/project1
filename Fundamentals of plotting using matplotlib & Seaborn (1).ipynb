**Elements in a plot:**
Figure: It is the area that holds all the visualization required
Title : A short description (not more than one line)
Plots : A chart that has some visualization
Sub-Plot : Also a chart but that is linked to another chart
Axis: How data gets represented. N-dimensional plot, has N-axis 2D: X, Y 3D: X,Y,Z
Axis-label : This is mostly the variable unit or variable definition that is being plotted
Legend : This is the representation of data.Categorical data in nature within any column of dataset
Gridlines: Checkboxes, used to easily interpret the values of axis on the plot
datatips or datalabels : These are data values on top of the plot

```python
x=list(range(10,100,10))
y=list(map(lambda x:pow(x,3),x))
z=list(map(lambda x:pow(x,2),x))
plt.plot(x,y,color='green', marker='*', linestyle='dashed',linewidth=3, markersize=12)
plt.plot(x,z,color='red', marker='o', linestyle='dashed',linewidth=3, markersize=12)
plt.xlabel('Number Scale')
plt.ylabel('Number Output Scale')
plt.title('Relation between number and its cube and Square')
plt.legend(['Power3', 'Power2'])
```

```python
x=list(range(10,100,10))
y=list(map(lambda x:pow(x,3),x))
z=list(map(lambda x:pow(x,2),x))
#fmt option for plotting matplot lib
#fmt = '[marker][line][color]'
#Eg [square marker][dashed line][bluecolor]=''s-b''
plt.plot(x,y,'s-b',linewidth=3,markersize=12)
plt.plot(x,z,'*-r', markersize=12)
plt.xlabel('Number Scale')
plt.ylabel('Number Output Scale')
plt.title('Relation between number and its cube and Square')
plt.legend(['Power3', 'Power2'])
```

```python
# Load data into a Pandas dataframe
import seaborn as sns
flowers_df = sns.load_dataset("iris")
```

```python
flowers_df
```

```python
#sns.scatterplot(x=flowers_df.sepal_length, y=flowers_df.sepal_width)
sns.scatterplot(x=flowers_df.sepal_length, y=flowers_df.sepal_width)
```

```python
#hue is the representation within a plot using color, that is possible for a categorical column
sns.scatterplot(x=flowers_df.sepal_length, y=flowers_df.sepal_width,hue=flowers_df.species)
```

```python
# Another way of plotting scatter plot, if u have a dataframe avl
#sns.scatterplot(x=flowers_df.sepal_length, y=flowers_df.sepal_width,hue=flowers_df.species)
plt.title('Sepal Dimensions')
sns.scatterplot(x='petal_length', y='petal_width', hue='species',data=flowers_df)
```

#Histogram


```python
plt.title("Distribution of Sepal Width")
plt.hist(flowers_df.sepal_width)
```

```python
plt.title("Distribution of Sepal Width")
plt.hist(flowers_df.sepal_width,bins=20)
```

```python
setosa_df = flowers_df[flowers_df.species == 'setosa']
versicolor_df = flowers_df[flowers_df.species == 'versicolor']
virginica_df = flowers_df[flowers_df.species == 'virginica']
plt.hist(setosa_df.sepal_width, alpha=0.4, bins=np.arange(2, 5, 0.25))
plt.hist(versicolor_df.sepal_width, alpha=0.4, bins=np.arange(2, 5, 0.25))
plt.hist(virginica_df.sepal_width, alpha=0.4, bins=np.arange(2, 5, 0.25))
```

```python
plt.title('Stacked Distribution of Sepal Width')
plt.hist([setosa_df.sepal_width, versicolor_df.sepal_width, virginica_df.sepal_width], 
         bins=np.arange(2, 5, 0.25), 
         stacked=True);
plt.legend(['Setosa', 'Versicolor', 'Virginica']);
```

#Heat Map



```python
#First requirement for heatmap is to create a pivot
flights_df = sns.load_dataset("flights").pivot("month", "year", "passengers")
```

```python
flights_df
```

```python
# plot using a color palette
sns.heatmap(flights_df, cmap="YlGnBu")
plt.show()

sns.heatmap(flights_df, cmap="Blues")
plt.show()

sns.heatmap(flights_df, cmap="BuPu")
plt.show()

sns.heatmap(flights_df, cmap="Greens")
plt.show()
```

```python
#Annotation set to True
plt.title("No. of Passengers (1000s)")
sns.heatmap(flights_df, fmt="d", annot=True, cmap='YlGnBu');
```

```python
import matplotlib.pyplot as plt
import numpy as np

plt.style.use('_mpl-gallery-nogrid')


# make data
x = [1, 2, 3, 4]
colors = plt.get_cmap('Blues')(np.linspace(0.2, 0.7, len(x)))

# plot
fig, ax = plt.subplots()
ax.pie(x, colors=colors, radius=3, center=(4, 4),
       wedgeprops={"linewidth": 1, "edgecolor": "white"}, frame=True)

ax.set(xlim=(0, 8), xticks=np.arange(1, 8),
       ylim=(0, 8), yticks=np.arange(1, 8))

plt.show()
```

```python
#Pie Chart Example 2
# importing libraries
import matplotlib.pyplot as plt
import seaborn as sns

# declaring data
data = [44, 45, 40, 41, 39]
keys = ['Class 1', 'Class 2', 'CLass 3', 'Class 4', 'Class 5']

# define Seaborn color palette to use
palette_color = sns.color_palette('bright')

# plotting data on chart
plt.pie(data, labels=keys, colors=palette_color, autopct='%.0f%%')

# displaying chart
plt.show()
```
